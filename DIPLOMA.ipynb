##### 3. Парсинг

Проходимся по каждой из *4760* страниц списка стримеров, проводим процедуры извлечения кода с каждой веб-страницей. На этой странице получаем список всех номеров __nums__, список всех имен __names__ и список всех игр __games__. Далее проходим по этим спискам и добавляем каждый элемент в результирующий массив.
В процессе добавления убираем лишние символы и преобразуем номер в тип __int__ для более удобного использования.

```python
rates = [] #результирующий массив
for i in range(1,4760):
    url = f"https://streamersbase.ru/streamers?page={i}" #ссылка на страницу, меняется по её номеру
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text)
    nums = soup.find_all("div",attrs={"class":"col-md-1 col-1 num"}) #список номеров
    names = soup.find_all("div", attrs={"class":"col-lg-3 col-md-10 col-8 name"}) #список имен 
    games = soup.find_all("div", attrs={"class":"col-lg-3 col-md-6 col-sm-8 col-8 game"}) #список игр
    for j in range(len(nums)): #проходимся по списку
        num = nums[j].get_text()
        num = num.replace('#', '') #убираем лишние символы
        num = int(num) #преобразуем в int
        name = names[j].get_text().strip() #получаем имя стримера
        game = games[j].get_text().strip() #получаем имя игры
        game = game.replace('\r\n', '') #убираем лишние символы
        game = game.replace('Последний стрим', '')
        game = game.replace('  ', '') 
        tag = [num, name, game] #добавляем все три характеристики в один элемент
        rates.append(tag) #кладем в конец списка
rates
```

##### 4. Преобразуем получившийся массив в формат DataFrame

```python
import pandas as pd
```

```python
df = pd.DataFrame(rates, columns=['Номер','Никнейм','Название игры'],
                 index = range(1,len(rates)+1))
df.head(5)
df.tail()
```
